```python
#CREATION D'UNE FONCTION PERMETTANT DE CALCULER LA MATRICE DE PROBABILITE DES BUTS POUR UN MATCH DONNE
def matrix_score_proba(id_team_home , id_team_away):
    
    # 1/DETERMINATION DES LAMBDAS NECESSAIRES AU CALCUL DE LA LOI DE POISSON (en utilisant les estimations)
        #lambda_home : le nombre de buts moyen de l'equipe qui joue a domicile
        #lambda_away : le nombre de buts moyen de l'equipe qui joue à l'exterieur
        #np.exp(a) : appel de la fonction exponentielle de numpy 
    lambda_home=np.exp(res_glm.params[str(id_team_home)+'_att']-res_glm.params[str(id_team_away)+'_def']+res_glm.params['home'])
    lambda_away=np.exp(res_glm.params[str(id_team_away)+'_att']-res_glm.params[str(id_team_home)+'_def'])
    
    # 2/BOUCLE PERMETTANT DE GENERER LES PROBAS POUR CHAQUE BUT ( proba de marquer 0 but, 1 but, 2 buts, 3 buts, 4 buts, 5 et plus)
        # buts marqués par l'équipe qui joue à domicile
    score_a=[] # declaration du tableau
    for i in range(5) : #boucle pour i=0 à 4
        score_a.append(poisson.pmf(i,lambda_home)) #proba pour les buts de 0 à 4 (poisson.pmf effectue la formule de Poisson)
    score_a.append(1-sum(score_a))#proba de marquer 5 buts ou plus
        # buts marqués par l'équipe qui joue à l'extérieur
    score_b=[]
    for i in range(5) :
        score_b.append(poisson.pmf(i,lambda_away))
    score_b.append(1-sum(score_b))
    
    # 3/CREATION DE LA MATRICE DES PROBABILITES DE BUTS
    A=np.array(score_a).reshape(6,1)#transposition du vecteur ligne en vecteur colonne
    B=np.array(score_b).reshape(1,6)
    
    return np.around(np.multiply(A,B)*100,2) #genere le tableaux de matrice de buts en pourcentage
```

```python
#BOUCLE QUI CALCULE LES MATRICES POUR LES 20 DERNIERS MATCHS (ECHANTILLON TEST)
mats=[] #declaration du tableau en sorti
for row in rows.index[-20:] : #boucledu 1er au 20eme match
    #print rows.ix[row,'name_team_home'] +'  vs   ' +rows.ix[row,'name_team_away']
    mats.append(matrix_score_proba(rows.ix[row,'id_team_home'],rows.ix[row,'id_team_away']))# Application de la fonction pour ts les matchs
```

```python
#Calcul des probabilités totales W D L
def prob_win_loose(mat_proba) :
    win = 0 #initialisation de la proba 
    draw = 0 #initialisation de la proba 
    loose = 0 #initialisation de la proba 
    for i in range(6): 
        for j in range(6):
            if (i > j) : #proba de gagner
                win = win + mat_proba[i,j]
            if (i == j ) : #proba de nul
                draw =draw + mat_proba[i,j]
            if (i < j) : #proba de perdre 
                loose = loose + mat_proba[i,j]
    return win,draw,loose
```
